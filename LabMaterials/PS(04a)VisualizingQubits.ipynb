# Global and Relative Phase:

We have earlier seen the action of $Z$ gate on $\ket{1}$. $Z$ gate flips the sign of $\ket{1}$. Now if you do a measurement after the operation $Z\ket{1}$ the measurement results would give a probability of nearly $1$ for the state $\ket{1}$. But the actual outcome state is $-\ket{1}$ and not $\ket{1}$.

When we measure $ \ket{v} $ or $ -\ket{v} $, we have the same statistics, and so we cannot know if it is $ \ket{v} $ or $ -\ket{v} $. They are indistinguishable (although they are not the same vectors). Thus, the set $ \{ \ket{v},-\ket{v} \} $ represents a single quantum state. And, we say that $ -\ket{v} $ differs from $ \ket{v} $ by a _global phase_ $-1$.


In general, $ e^{i\phi} \ket{v} $ differs from $ \ket{v} $ by a global phase $ e^{i\phi} $.

**Thus, $ \{ e^{i\phi} \ket{v} \mid 0 \leq \phi < 2\pi \} $ represents a single quantum state.**

<font color='blue'> **A global phase does not affect the computation and so the outcomes. It can thus be removed.**</font>

## How to determine if two states differ by a global phase or not?

**<font color='yellow'> Having the same measurement outcomes is a necessary condition but not sufficient.</font>**

Consider the states $ \ket{+} $ and $ \ket{-} $. The probabilities of observing the basis states $ \ket{0} $ and $ \ket{1} $ are $ \dfrac{1}{2} $ for both $ \ket{+} $ and $ \ket{-} $, but they are perfectly distinguishable states. If we apply Hadamard and then measure, we observe only $ \ket{0} $ (or $\ket{1}$) for $ \ket{+} $ (or $\ket{-}$).

##### <font color='red'> Using inner product to determine if two states differ by a global phase</font>

Let's say,  $ \ket{v} $ differs from $ \ket{u} $ by a global phase $ e^{i\theta} $ (which is non-zero), that is $\ket{v} = e^{i\theta} \ket{u}$

We know that $ \braket{u|u} = 1  $ if $\ket{u} $ is a quantum state. We now have, 

$ \braket{u|v} = \braket{u|e^{i\theta}u} = e^{i\theta} \braket{u|u} = e^{i\theta} $ (which is non-zero)

**<font color='yellow'>Thus in addition to same measurement outcomes, we have a nonzero inner product for two states differing by a global phase.</font>**

## How to uniquely represent a quantum state? 

As $ \{ e^{i\phi} \ket{v} \mid 0 \leq \phi < 2\pi \} $ represents a single quantum state how to uniquely represent the state $\ket{v}$ ? Another way to ask this question is, what is the quantum state that is representative of the infinitely many possible states $ e^{i\phi} \ket{v} $

<font color='yellow'>**Rule: The amplitude of $ \ket{0} $ in a quantum state shall always be a real number.**</font>

Consider the state $ \ket{u} = \dfrac{1+i}{2}\ket{0} + \dfrac{1-i}{2} \ket{1} $.

This is simply,  $ \ket{u} = \frac{1}{\sqrt{2}} e^{i \pi \over 4} \ket{0} +  \frac{1}{\sqrt{2}} e^{- i \pi \over 4} \ket{1} $ (show this, it is a simple exercise on complex numbers !!)

By factoring out $ e^{i \pi \over 4} $, the amplitude of $ \ket{0} $ will be a real number.

$ \ket{u} = e^{i \pi \over 4} ( { \frac{1}{\sqrt{2}} \ket{0} + \frac{1}{\sqrt{2}} e^{-i \pi / 2 } \ket{1} } ) $

As the length of $ e^{i \pi \over 4} $ is 1, it is a global phase and we can remove it. Thus, we have $ \ket{u'} = \frac{1}{\sqrt{2}} \ket{0} + \frac{1}{\sqrt{2}} e^{-i \pi / 2 } \ket{1}  $  as the unique representation of $\ket{u}$.
 
($ \ket{u} $ differs from $ \ket{u'} $ by a global phase $ e^{i \pi \over 4} $, hence we consider them to be indistinguishable states)

**Using the above rule, unique representation of a state is given by:**

##### $ r_1 \ket{0} + r_2 e^{i\phi} \ket{1},  $

where
- $ r_1 $ and $ r_2 $ are non-negative real numbers,
- $ r_1^2 + r_2^2 = 1 $, and
- $ |e^{i\phi}| = 1 $.
- The angle $\phi$ is called the relative phase

## <font color='brown'> Visualization of quantum states </font>

By using the unique representation, each quantum state can be visualized on three dimensions.
The unique representation has three parameters $ r_1 $, $ r_2 $, and $ \phi $. We combine these and use them to visualize the quantum states of a single qubit.



- $\{ e^{i\phi} \} $ forms a unit circle on the complex plane ( where relative phase $\phi \in [0,2\pi) $ ).

<img src="Images/phase.png" alt="Phase figure" width="300"/>

- Also $ r_1^2 + r_2^2 = 1 $ and both $ r_1 $ and $ r_2 $ are non-negative real numbers. Thus, all $ (r_1,r_2) $ pairs form the first quadrant of the unit circle.

<img src="Images/r_s.png" alt="Real Amplitudes" width="300"/>


You can see from the figure above that, we can represent any pair $ (r_1,r_2) $ by its angle $ \theta \in [0,\pi /2] $, the angle with the $ \ket{0} $-axis:

- $ r_1 = \cos(\theta) $
- $ r_2 = \sin(\theta) $

##### **Thus, we require only $2$ real variables $(\theta,\phi)$ to specify the state of a qubit.**

### Bloch Sphere

<font color='yellow'> We use a unit sphere to visualize </font> $((r1,r2),\phi)$ <font color='yellow'> or equivalantly </font> $(\theta,\phi)$.

- We can place the unit circle due to $ \phi $ at the equator.

- Then, we can use a semi-sphere to place the first-quadrant into the picture. But, then, we obtain a non-symmetric object. The extreme points of $ \theta $ are 0 and $ \pi/2 $, and there is only a single pole to place one of these extreme points. One practical solution is streching the angle of magnitudes by a factor of 2 and so using both poles.

    - We define $ \theta \in [0,\pi] $ and place $ \ket{0} $ and $\ket{1}$ at the top (north pole) and bottom (south pole) respectively. 

    - Thus, given $ \theta $ as above, the magnitudes should be calculated as:

        - $ r_1 = \cos ( \dfrac{\theta}{2} ) $

        - $ r_2 = \sin ( \dfrac{\theta}{2} ) $

Thus, the unique representation of a one qubit state can be written in terms of $ (\theta,\phi)  $ as:

#### $ \ket{u} = \cos ({\frac{\theta}{2}}) \ket{0} + e^{i\phi} \sin ({\frac{\theta}{2}})\ket{1} $.

We now have all the ingredients to visualize a one-qubit state. The last step is to combine the semi-circle due to $ \theta $ with the unit circle due to the relative phase $ \phi $.


We use $ \dfrac{ \ket{0}+\ket{1} }{ \sqrt{2} } = \ket{+}$ as a reference point. Its relative phase ($\phi$) is $0$ because $ e^{i\phi} = 1 $ for $\ket{+}$.

On the left, we see any quantum state on the Bloch sphere having the unique representation
$ \ket{\psi} = \cos ({\frac{\theta}{2}}) \ket{0} + e^{i\phi} \sin ({\frac{\theta}{2}}) \ket{1} $.
- Observe that the reference axis for the angles are the $ z $-axis and $ x $-axis.

On the right, we see the Bloch sphere with the six reference quantum states:
- $ \ket{0} $ and $ \ket{1} $ are placed at the intersection of $z$-axis with the unit sphere, where $ \ket{0} $ is on the positive side (North Pole).
- $ \ket{+} $ and $ \ket{-} $ are placed at the intersection of $x$-axis with the unit sphere, where $ \ket{+} $ is on the positive side.
- $ \ket{i} $ and $ \ket{-i} $ are placed at the intersection of $y$-axis with the unit sphere, where $ \ket{i} $ is on the positive side.

## Orthogonal states on a Bloch Sphere

Let $ \ket{u} = \cos ({\frac{\theta}{2}}) \ket{0} + e^{i\phi} \sin ({\frac{\theta}{2}})\ket{1} $

Then, the quantum state on the opposite side on the Bloch sphere is

$ \ket{v} = \sin ({ \frac{\theta}{2} }) \ket{0} +  e^{i(\phi+\pi)} \cos ({ \frac{\theta}{2} }) \ket{1}) $

The above state was obtained by changing $ ({\theta, \phi}) $ in $\ket{u}$ to $ ({\pi-\theta,\phi+\pi}) $.

Observe that:
- $ \cos ({ \frac{\pi-\theta}{2} }) = \cos ({ \frac{\pi}{2} - \frac{\theta}{2} }) = \sin ({\frac{\theta}{2}}) $ and

- $ \sin ({ \frac{\pi-\theta}{2} }) = \sin ({ \frac{\pi}{2} - \frac{\theta}{2} } ) = \cos ( {\frac{\theta}{2}} )$.


**Explicitly show that the states $\ket{u}$ and $\ket{v}$ are indeed orthogonal states by showing $ \braket{u|v}=0 $**
